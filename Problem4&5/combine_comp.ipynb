```python
#Importing Libraries
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

#for themes
from jupyterthemes import jtplot
jtplot.style(theme = "monokai",context = "notebook", ticks = True,grid = False)
```

```python
#Declaring path variable
path = '/Users/kushagrakhatwani/MDS/Project1/Problem4&5' #Set your own path for the path variable

#Importing red part of the reconstructed image
r = np.load(path+"/R/Red.npy")
#print(r)
```

```python
#Importing blue part of the reconstructed image
b = np.load(path+"/B/Blue.npy")
#print(b)
```

```python
#Importing green part of the reconstructed image
g = np.load(path+"/G/Green.npy")
#print(g)
```

```python
#Combing all components into one RGB image
im = np.zeros((100,100,3))
print(im.shape)

#Normalizing images so all pixel intensity values lie between 0-255
im[:,:,0] = (r-np.min(r))/(np.max(r)-np.min(r))
im[:,:,1] = (g-np.min(g))/(np.max(g)-np.min(g))
im[:,:,2] = (b-np.min(b))/(np.max(b)-np.min(b))

#print(im[20,20,:])
#print(im)
```

```python
#Plotting reconstructed image
plt.imshow(im)
plt.imsave('reconstructed.png',im)
```

```python
#Comparing with incomplete image
plt.figure()
#Importing incomplete image
orig = Image.open(path+'/img.png')
plt.subplot(1,2,1)
plt.imshow(orig)
plt.title("Original")

plt.subplot(1,2,2)
plt.imshow(im,cmap='gray')

#Reconstruction when corruption 70%
plt.title("Reconstructed")
```
